```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Distribution 1: Pareto

```python
pareto = np.random.pareto(5, 100)
plt.hist(pareto, color='c')
plt.axvline(pareto.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(pareto.mean() + pareto.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(pareto.mean() - pareto.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

The descriptive statistics are not useful here because the distribution is heavily skewed.

Distribution 2: Negative binominal

```python
neg_bi = np.random.negative_binomial(10, 0.75, 100)
plt.hist(neg_bi, color='c')
plt.axvline(neg_bi.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(neg_bi.mean() + neg_bi.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(neg_bi.mean() - neg_bi.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

The descriptive statistics are better here than for the pareto but still do not capture the slight skew.

Distribution 3: Chi-square

```python
chi_sq = np.random.chisquare(5, 100)
plt.hist(chi_sq, color='c')
plt.axvline(chi_sq.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(chi_sq.mean() + chi_sq.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(chi_sq.mean() - chi_sq.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

Again the descriptive statistics do not capture the skew of the distribution.

Distribution 4: Exponential

```python
exp = np.random.exponential(10, 100)
plt.hist(exp, color='c')
plt.axvline(exp.mean(), color='b', linestyle='solid', linewidth='2')
plt.axvline(exp.mean() + exp.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(exp.mean() - exp.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

The skew is not captured here either.

Distribution 5: Lognormal

```python
log_n = np.random.lognormal(10, 10, 100)
plt.hist(log_n, color='c')
plt.axvline(log_n.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(log_n.mean() + log_n.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(log_n.mean() - log_n.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

The statistics are useless here.

Distribution 6: Uniform

```python
uniform = np.random.uniform(10, 20, 100)
plt.hist(uniform, color='c')
plt.axvline(uniform.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(uniform.mean() + uniform.std(), color='b', linewidth=2, linestyle='dashed')
plt.axvline(uniform.mean() - uniform.std(), color='b', linewidth=2, linestyle='dashed')
plt.show()
```

The statistics are far more useful here.

The distributions that are usefully described by the statistics are the symmetric ones. Those with a skew are badly described.

Additionally:

```python
varb1 = np.random.normal(5, 0.5, 100)
varb2 = np.random.normal(10, 1, 100)
varb3 = varb1 + varb2
plt.hist(varb3, color='c')
plt.axvline(varb3.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(varb3.mean() + varb3.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(varb3.mean() - varb3.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

The descriptive statistics are fairly good here. There is potentially a slight skew here, but it is hard to tell whether this is a result of the sampling and the bins.
